```python
# Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Q7) Look at the data given below. Plot the data, find the outliers and find out  μ,σ,σ^2

```python
data=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
title=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
# Mean,Variance and Std_dev of the dataset

mean = data.mean() ; print('mean = {}'.format(mean));
variance = data.var() ; print('variance = {}'.format(variance));
std_dev = data.std() ; print('std_dev = {}'.format(std_dev));
```

```python
# Box Plot to find outliers
sns.boxplot(x =data)
```

Outliers are present in the dataset

```python
# Histogram to find skewness
sns.histplot(data)
```

The data is right skewed, i.e extreme values are present on the right side

```python
# Pie Plot to get distribution of dataset
figure = plt.figure(figsize = (11,8))
plt.pie(data,labels=title,autopct='%1.0f%%')
plt.show
```
