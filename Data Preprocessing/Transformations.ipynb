```python
# Histogram of all the features
fig,axs = plt.subplots(1,2)
sns.histplot(data['Weight gained (grams)'],ax = axs[0])
sns.histplot(data['Calories Consumed'], ax = axs[1])
```

```python
# histogram and QQ plots

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot((data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot((data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot((data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot((data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

```python
# log transformation on features

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot(np.log(data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot(np.log(data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot(np.log(data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot(np.log(data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

```python
# Square transformation on features

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot(np.square(data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot(np.square(data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot(np.square(data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot(np.square(data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

```python
# Square root transformation on features

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot(np.sqrt(data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot(np.sqrt(data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot(np.sqrt(data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot(np.sqrt(data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

```python
# Reciprocal transformation on features

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot(1/(data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot(1/(data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot(1/(data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot(1/(data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

```python
# Using Box Cox transformation
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='box-cox',standardize = False)

pt_data = pt.fit_transform(data)
pt_data = pd.DataFrame(pt_data, columns=data.columns)
```

```python
# Checking the values of lambda for each feature
pd.DataFrame({'cols':data.columns,'box-cox-lambdas':pt.lambdas_})
```

```python
# Histogram and QQ plot for box-cox transformed data

plt.figure(figsize=(10,6))
plt.subplot(121)
sns.histplot((pt_data['Weight gained (grams)']),bins = 4,kde =1)
plt.subplot(122)
stats.probplot((pt_data['Weight gained (grams)']),dist='norm',plot=plt)
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(221)
sns.histplot((pt_data['Calories Consumed']),kde =1)
plt.subplot(222)
stats.probplot((pt_data['Calories Consumed']),dist='norm',plot=plt)

plt.show()
```

Box-Cox Transformation gives the best result. The data now is normally distributed.
