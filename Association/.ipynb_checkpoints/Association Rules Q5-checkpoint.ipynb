# Association Rules : Transactions Retail Database

What is business objective?
    Business objective of this dataset is to check which products are bought together 

Constraints: Combinations of products limited

```python
# importing necessary libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
# importing the Transaction Retail dataset

products = []
with open('C:\\Users\\vaibh\\Desktop\\360 Digitmg\\Association\\transactions_retail1.csv') as f:
    products = f.read()
first_split = products.split('\n')
products_list = []

for i in first_split:
    products_list.append(i.split(','))
products_list
```

```python
# Removing NA values from the dataset

Cleaned_products = []
for j in range(len(products_list)):
    temp = []
    for i in products_list[j]:
        i = i.replace("'",'')
        if i != 'NA':
            temp.append(i)
    temp = ' '.join(temp)
    Cleaned_products.append(temp)
Cleaned_products
```

```python
# Frequency of each product

from collections import Counter
item_frequency = Counter(Cleaned_products)

# Sorting in reverse the frequency 
item_frequency = sorted(item_frequency.items(),key = lambda x:x[1],reverse = True)
item_frequency
```

```python
# Separating frequency and items from item_frequency
items = [item[0] for item in item_frequency]
frequency = [item[1] for item in item_frequency] 
```

```python
# Barplot of most sold items
sns.barplot(frequency[:11],items[:11])
```

    Tins are the most sold product with frequency of 4277.

```python
# Converting List to Dataframe 

New_products = pd.DataFrame(pd.Series(Cleaned_products),columns = ['Products'])

New_products.head(15)
```
