# Association Rules : Groceries Database

What is business objective?
    Business objective of this dataset is to increase the sales of groceries.

Constraints: Number of product combinations limited.

```python
# importing necessary libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
# importing the groceries dataset
groceries = []
with open('C:\\Users\\vaibh\\Desktop\\360 Digitmg\\Association\\groceries.csv') as f:
    groceries = f.read()
first_split = groceries.split('\n')
groceries_list = []

for i in first_split:
    groceries_list.append(i.split(','))
groceries_list
```

```python
# Frequency of each item in groceries list
all_groceries = [i for item in groceries_list for i in item]

from collections import Counter
item_frequency = Counter(all_groceries)

# Sorting in reverse the frequency 
item_frequency = sorted(item_frequency.items(),key = lambda x:x[1],reverse = True)
item_frequency
```

```python
# Separating frequency and items from item_frequency
items = [item[0] for item in item_frequency]
frequency = [item[1] for item in item_frequency] 
```

```python
# Barplot of most sold items
sns.barplot(frequency[:11],items[:11])
```

Whole milk is the most sold item with frequency of 2513, followed by other vegetables with frequency of 1903

```python
# Converting List to Dataframe 
groceries_series = pd.DataFrame(pd.Series(groceries_list),columns = ['transactions'])
groceries_series = groceries_series.iloc[:-1]
groceries_series.head()
```

```python
# Creating dummy variables for each item in transactions
dataset = groceries_series['transactions'].str.join(sep='*').str.get_dummies(sep='*')
dataset.head()
```

```python
# Importing apriori and association rules

from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(dataset,min_support = 0.0075,use_colnames = True, max_len = 4)
frequent_itemsets = frequent_itemsets.sort_values('support',ascending = False)
frequent_itemsets
```

```python
# Barplot of the supports for first 12 itemsets 

sns.barplot(y = frequent_itemsets['support'][0:11],x = frequent_itemsets['itemsets'][0:11])
plt.xticks(rotation = 80)
plt.show()
```

```python
# Association rules of the dataset

rules = association_rules(frequent_itemsets,metric = 'lift',min_threshold = 1)
rules = rules.sort_values('lift',ascending=False)
rules.head()
```

```python
# Removing redundant itemsets
def to_list(i):
    return sorted(list(i))

red_data = rules.antecedents.apply(to_list) + rules.consequents.apply(to_list)
red_data = red_data.apply(sorted)
rules_sets = list(red_data)
rules_sets
```

```python
# Getting unique rules 
unique_rules_set = [list(m) for m in set(tuple(i) for i in rules_sets)]
unique_rules_set
```

```python
# Retrieving index of unique itemsets
index_rules = []

for i in unique_rules_set:
    index_rules.append(rules_sets.index(i))
rules_no_red = rules.iloc[index_rules,:]
rules_no_red = rules_no_red.sort_values('lift',ascending= False)
rules_no_red
```

```python
# Scatter plot 
sns.scatterplot(x = rules_no_red.support, y = rules_no_red.confidence, hue=rules_no_red.lift,alpha=0.8)
```

Observations:
1. In the dataset, 409 rules were found using apriori algorithm
2. More than 50% data is found in the support level of 0.005 to 0.015
3. Lift ratio ranges from 1 to 4.15
4. The highest lift ratio corresponds to confidence value of 0.28 and support value of 0.008
