```python
# Checking for outliers using boxplot

fig, axs = plt.subplots(ncols=3, nrows=3, figsize=(12, 10))
index = 0
axs = axs.flatten()
data1 = Num_data.iloc[:,0:]
for k,v in data1.items():
    sns.boxplot(k, data=data1, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.55, w_pad=0.5, h_pad=4.0)
```

```python
# Winsorizing the outliers to retain the data from the outliers

for col in Num_data.columns:
    IQR = Num_data[col].quantile(0.75) - Num_data[col].quantile(0.25)
    upper_end = Num_data[col].quantile(0.75) + (1.5*IQR)
    low_end = Num_data[col].quantile(0.25) - (1.5*IQR)
    if low_end < 0:
        low_end = 0
        Num_data[col] = pd.DataFrame(np.where(Num_data[col] > upper_end,upper_end, 
                          np.where(Num_data[col] < low_end,low_end,Num_data[col])))
    else:
        Num_data[col] = pd.DataFrame(np.where(Num_data[col] > upper_end,upper_end, 
                          np.where(Num_data[col] < low_end,low_end,Num_data[col])))        
```

```python
# Importing Variance Threshold for Zero Variance
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.0)
vt_data = vt.fit(Num_data)
vt_data.get_support()
```

```python
# Dropping Zero Variance features
Num_data.drop(['Number of Open Complaints'], axis=1,inplace = True)
```

```python
# Converting Categorical Data into Numerical Data
# Label Encoding
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
lb_data_a = lb.fit_transform(Cat_data['Response'])
lb_data_b = lb.fit_transform(Cat_data['Gender'])
lb_data_A = pd.DataFrame(lb_data_a)
lb_data_B = pd.DataFrame(lb_data_b)
Cat_data['Coverage'] = Cat_data['Coverage'].map({'Basic':0,'Extended':1,'Premium':2})
Cat_data['Education'] = Cat_data['Education'].map({'High School or Below':0,'College':1,'Bachelor':2,'Master':3,'Doctor':4})
Cat_data['Location Code'] = Cat_data['Location Code'].map({'Rural':0,'Suburban':1,'Urban':2})
Cat_data['Marital Status'] = Cat_data['Marital Status'].map({'Divorced':0,'Single':1,'Married':2})
Cat_data['Policy Type'] = Cat_data['Policy Type'].map({'Corporate Auto':0,'Personal Auto':1,'Special Auto':2})
Cat_data['Renew Offer Type'] = Cat_data['Renew Offer Type'].map({'Offer1':0,'Offer2':1,'Offer3':2,'Offer4':3})
Cat_data['Vehicle Size'] = Cat_data['Vehicle Size'].map({'Small':0,'Medsize':1,'Large':2})
```

```python
Cat_data.head()
```

```python
# Converting Categorical(Nominal) data to numerical using getdummies
list1 = ['State','EmploymentStatus','Sales Channel','Vehicle Class']
dum = pd.DataFrame([])
for feat in list1:
    cat_dum = pd.get_dummies(data[feat],prefix=feat,drop_first = True)
    dum = pd.concat([dum,cat_dum],axis=1)
```

```python
# Droppping features
Cat_data.drop(['State','EmploymentStatus','Sales Channel','Vehicle Class','Policy','Response','Gender'],axis = 1, inplace=True)
```

```python
# Concatening the categorical features
New_Cat_data = pd.concat([Cat_data,dum,lb_data_A,lb_data_B],axis=1)
```

```python
New_Cat_data.head()
```

```python
# Concatening Numerical and Categorical data

Trans_data = pd.concat([Num_data,New_Cat_data],axis = 1)
Trans_data
```

# Standardization and Normalization

```python
# Standardization 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
std_data = pd.DataFrame(sc.fit_transform(Trans_data),columns = Trans_data.columns)
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
norm_data = pd.DataFrame(mm.fit_transform(Trans_data),columns = Trans_data.columns)
```

# KMeans Clustering 

```python
# K-Means Clustering to check for number of clusters on standardised dataset
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,10):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(std_data)
    wcss.append(kmeans.inertia_)
```

```python
# Visualizing the Elbow method to get the optimal value of K value

plt.plot(range(1,10), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# K-Means Clustering to check for number of clusters on normalized dataset
from sklearn.cluster import KMeans

wcss2=[]

for i in range(1,10):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(norm_data)
    wcss2.append(kmeans.inertia_)
```

```python
# Visualizing the Elbow method to get the optimal value of K value

plt.plot(range(1,10), wcss2)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Checking silhoeutte score for normalized and standardized data

from sklearn.metrics import silhouette_score
print('\nFor Standardized data')
for cluster in range(2,6):
    temp = KMeans(n_clusters = cluster, init='k-means++', random_state=0)
    cluster_labels = temp.fit_predict(std_data)
    s_score = silhouette_score(std_data,cluster_labels)
    print('For {} clusters silhouette score is {}.'.format(cluster,s_score))

print('\nFor Normalized data')
for cluster in range(2,6):
    temp = KMeans(n_clusters = cluster, init='k-means++', random_state=0)
    cluster_labels = temp.fit_predict(norm_data)
    s_score = silhouette_score(norm_data,cluster_labels)
    print('For {} clusters silhouette score is {}.'.format(cluster,s_score))
```

    Using Normalized data with 4 clusters

```python
# Predicting clusters using Kmeans

kmeansmodel = KMeans(n_clusters= 4, init='k-means++', random_state=0)
cluster_labels = kmeansmodel.fit_predict(norm_data)
```

```python
# Adding cluster labels to original dataset

Trans_data['cluster_labels'] = pd.DataFrame(cluster_labels) 
```

```python
# Countplot of the clusters
sns.countplot(Trans_data.cluster_labels)
```
