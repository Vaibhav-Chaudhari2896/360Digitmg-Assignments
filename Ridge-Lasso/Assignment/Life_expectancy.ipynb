```python
# Checking for columns with null values
null_col = []
for i in data.columns:
    if data[i].isnull().sum() != 0:
        null_col.append(i)

# Boxplot for columns with null values
fig, axes = plt.subplots(4,4, figsize=(15,10))
index = 0
axes = axes.flatten()

for col in null_col: 
    sns.boxplot(x = data[col], ax = axes[index])
    index += 1
plt.tight_layout(w_pad = 0.5, h_pad = 0.5, pad = 1) 
```

    Apart from Alcohol and BMI all other features have outliers. Using median for features with outliers and mean for without to fill missing values in the dataset

```python
# Replacing null values 
null_col
for i in null_col:
    if i == 'Alcohol' or 'BMI':
        data[i] = data[i].fillna(data[i].mean())
    else:
        data[i] = data[i].fillna(data[i].median())
```

```python
# Droping irrelevant features
data.drop(['Country','Year'], axis = 1, inplace=True)
```

```python
data.shape
```

```python
# Converting categorical to numerical 
data.Status = data.Status.map({'Developing':0, 'Developed':1})
```

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.1)
data = vt.fit_transform(data)
data = pd.DataFrame(data, columns = vt.get_feature_names_out())
data.head()
```

```python
# Boxplot of features
fig, axes = plt.subplots(4,5, figsize=(10,5))
index = 0
axes = axes.flatten()
for col in data.columns: 
    sns.boxplot(x = data[col], ax = axes[index])
    index += 1
plt.tight_layout(w_pad = 0.5, h_pad = 0.5, pad = 1)    
```

```python
for col in data.columns[1:]:
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    low_end = data[col].quantile(0.25) - (1.5*IQR)
    upper_end = data[col].quantile(0.75) + (1.5*IQR)
    if low_end < 0:
        low_end = 0
        data[col] = np.where(data[col] < low_end, low_end, data[col])
        data[col] = np.where(data[col] > upper_end, upper_end, data[col])
    else:
        data[col] = np.where(data[col] < low_end,low_end,data[col])
        data[col] = np.where(data[col] > upper_end,upper_end,data[col])
```

```python
# Correlation between features
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

# MODEL BUILDING

```python
# Splitting into x and y
x = data.drop('Life_expectancy', axis = 1)
y = data.Life_expectancy
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
# Normalizing
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms_x_train = mms.fit_transform(x_train)
mms_x_test = mms.transform(x_test)
```

```python
mms_x_train.shape, mms_x_test.shape, y_train.shape, y_test.shape
```

# Simple Linear Regression

```python
# Simple Linear Regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr_model = lr.fit(mms_x_train,y_train)  # fitting to train data
```

```python
pred_lr_test = lr_model.predict(mms_x_test) # predicting using test data
```

```python
from sklearn.metrics import r2_score
r2_score_lr = r2_score(y_test, pred_lr_test)
r2_score_lr
```

```python
# RMSE for linear regression

resid_lr = (pred_lr_test-y_test)
rmse_lr = np.sqrt(np.mean(resid_lr*resid_lr))
rmse_lr
```

# Lasso Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
lasso = Lasso()
parameters = ({'alpha': np.arange(0.1,5,0.1)})

lasso_grid = GridSearchCV(lasso, parameters, scoring='r2', cv = 5)
lasso_grid.fit(mms_x_test,y_test)
print(lasso_grid.best_params_)
print(lasso_grid.best_score_)
```

```python
lasso1 = Lasso(alpha=0.1)

lasso_model = lasso1.fit(mms_x_train,y_train)
```

```python
pred_lasso_test = lasso1.predict(mms_x_test)
```

```python
r2_score_lasso = r2_score(pred_lasso_test,y_test)
r2_score_lasso
```

```python
# RMSE for lasso regression

resid_lasso = (pred_lasso_test - y_test)
rmse_lasso = np.sqrt(np.mean(resid_lasso*resid_lasso))
rmse_lasso
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 0.1)
ridge_model = ridge.fit(mms_x_train,y_train)
```

```python
pred_ridge_test = ridge_model.predict(mms_x_test)
```

```python
r2_score_ridge = r2_score(y_test,pred_ridge_test)
r2_score_ridge
```

```python
# RMSE for ridge regression

resid_ridge = (pred_ridge_test - y_test)
rmse_ridge = np.sqrt(np.mean(resid_ridge*resid_ridge))
rmse_ridge
```

```python
results = pd.DataFrame([])
results['Regression'] = ['Simple', 'Lasso', 'Ridge']
results['R2_score'] = [r2_score_lr, r2_score_lasso, r2_score_ridge]
results['RMSE'] = [rmse_lr, rmse_lasso, rmse_ridge]
results
```

    Simple Regression gives better results than ridge and lasso regression
