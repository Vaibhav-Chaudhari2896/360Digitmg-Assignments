```python
# Using Kaplan Meier
from lifelines import KaplanMeierFitter

T = data.survival_time_hr    # Time Event
C = data.alive               # Censored event

kmf = KaplanMeierFitter()
kmf.fit(T, C, label='Kaplan Meier Estimate')
```

```python
kmf.plot()
plt.show()
```

```python
# Checking survival curves with respect to different groups

plt.figure(figsize=(12,8))
ax = plt.subplot(111)
kmf = KaplanMeierFitter()

T = data.survival_time_hr    # Time Event
C = data.alive       

for g in data.group.unique():
    temp = data.group == g
    kmf.fit(T[temp], event_observed = C[temp], label = g)
    kmf.plot(ax=ax)

plt.title("Survival curves by groups");
```

Clearly survival probability for group 3 is the best compared to group 1 and 2 upto 35 hrs. Beyond this Group 2 has better chance of survival

# Log Rank Test

```python
from lifelines.statistics import logrank_test

data.loc[(data.alive == 1) & (data.survival_time_hr <= 24), 'censored'] = 1
data.loc[(data.alive == 1) & (data.survival_time_hr > 24), 'censored'] = 0
data.loc[data.alive == 0, 'censored'] = 0
C2 = data['censored']

T_low = T[data.age < data.age.mean()]
T_high = T[data.age > data.age.mean()]
C_low = C2[data.age < data.age.mean()]
C_high = C2[data.age > data.age.mean()]

results = logrank_test(T_low, T_high, event_observed_A = C_low, event_observed_B = C_high)
results.print_summary()
```

p-value is greater than 0.05%. There is no significant difference in the survival time for people with age below 62 and above 62 for the first 24hrs

# Cox proportional hazards model

```python
from lifelines import CoxPHFitter
cph = CoxPHFitter()   # instantiating the model

cph.fit(data, 'survival_time_hr', event_col='alive')

cph.print_summary()
```

```python
# Predicting survival probability for each row
cph.predict_survival_function(data)
```

```python
# PLotting coefficient of each variable
cph.plot()
```
