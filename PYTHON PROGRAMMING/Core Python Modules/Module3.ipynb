## Statsmodels

```python
####################### Statsmodels##################################
##################Linear Regression Models######################
#################Ordinary Least Squares#########################
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std
np.random.seed(9876789)
# OLS estimation
nsample = 100
x = np.linspace(0, 10, 100)
X = np.column_stack((x, x**2))
beta = np.array([1, 0.1, 10])
e = np.random.normal(size=nsample)
#Our model needs an intercept so we add a column of 1s
X = sm.add_constant(X)
y = np.dot(X, beta) + e
# Fit and summary:
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

###  Handling duplicates and Categorical

```python
#Using set() Function to Remove Duplicate Items from list
#As we know that set is the unordered collection of unique elements. Python internally uses 
#the hash technique to process sets. It is quite simple and fast method to remove duplicate items 
#from Python list. The original order of the sequence (list) may not be preserved using this 
#approach. This method will also not work if you have dict or similar object in the list because 
#dict object is not hashable in Python.

#Example 1:
list1 = [1, 2, 3, 4, 1, 2, 3, 5, 6, 7]
print('List with duplicate elements: ',list1)
 
# Following code will remove the duplicate elements from list
list2 = list(set(list1))
 
# Printing the list with unique elements.
print('List with unique elements: ',list2)
```

```python
#Example 2:
mylist = ['a', 'a', 'b', 'b', 1, 2, 3, 4, 1, 2, 3, 5, 6, 7]
 
newlist1 = [i for n, i in enumerate(mylist) if i not in mylist[:n]]
 
newlist2 = [mylist[i] for i, x in enumerate(mylist) if x not in mylist[i+1:]]
 
print('newlist1:', newlist1)
print('newlist2:', newlist2)
```

```python
#Example 3:
#Using OrderedDict of collections Library to Remove Duplicate Items from list in Python.
from collections import OrderedDict
 
mylist = ['a', 'a', 'b', 'b', 1, 2, 3, 4, 1, 2, 3, 5, 6, 7]
 
# returns OrderedDict object
myListObj = OrderedDict.fromkeys(mylist)
print('myListObj:', myListObj)
 
# convert OrderedDict object to list
myListObj = list(myListObj)
print('myListObj:', myListObj)
 
# just a blank line
print()
 
# returns odict_keys object - only keys
myListObj = OrderedDict.fromkeys(mylist).keys()
print('myListObj:', myListObj)
 
# convert odict_keys object to list
myListObj = list(myListObj)
print('myListObj:', myListObj)
```

```python

####Demonstrating filter() to remove odd numbers

def filterOddNum(in_num):

    if(in_num % 2) == 0:
        return True
    else:
        return False
numbers= np.arange(10)
out_filter = filter(filterOddNum, numbers)

print("Type of filter object: ", type(out_filter))
print("Filtered seq. is as follows: ", list(out_filter))

#### A couple of points to notice in the example are:

####The filter() function is returning out_filter, and we used type() to check its data type.
####We called the list() constructor to convert the filter object to a Python list.

```

## Basic Image processing using Python OpenCV

```python
import cv2
import matplotlib.pyplot as plt
# load the example image
image = cv2.imread("download.jpg")
image.shape, type(image)
```

```python
plt.imshow(image)
```

```python
#resize the image
res = cv2.resize(image, dsize=(1000, 1000), interpolation=cv2.INTER_CUBIC) #dsize(width,height)
res.shape, plt.imshow(res)
```

```python
#convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
#blurring
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
plt.imshow(blurred)
```

```python
#edged
edged = cv2.Canny(blurred,50,200,250)
plt.imshow(edged)
```

### Face Capturing from vedio

```python
import cv2 

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml') 

def detect(gray, frame): 
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
    for (x, y, w, h) in faces: 
        cv2.rectangle(frame, (x, y), ((x + w), (y + h)), (255, 0, 0), 2) 
        roi_gray = gray[y:y + h, x:x + w] 
        roi_color = frame[y:y + h, x:x + w] 
    return frame 

video_capture = cv2.VideoCapture(0) 
while True:   
    _ , frame = video_capture.read()                            
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)     
    canvas = detect(gray, frame)                      
    cv2.imshow('Video', canvas)
    if cv2.waitKey(1)==ord('q'): #ord('q')--ASACII value of q
        break
    # Release the capture once all the processing is done. 
video_capture.release()                                  
cv2.destroyAllWindows() 

#The following is not working why ?-- It supposed to work if any key pressed inadditions to q.
#if cv2.waitKey(1):
 #       break
    
```
